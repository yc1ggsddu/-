```python
import pandas as pd
import numpy as np
```

```python
# https://www.kaggle.com/liujt14/dou-ban-movie-short-comments-10377movies
data=pd.read_csv("./archive/all_movies_with_id.csv",nrows=1000000,encoding='utf-8')
```

```python
data.columns
```

```python
data=data.drop(columns=['ID', 'Movie_Name', 'Score', 'Review_People', 'Star_Distribution',
       'Craw_Date', 'Username', 'Date', 
       'Comment_Distribution', 'Like'])
```

```python
data
```

```python
new_index=data.columns.insert(0,'label')
```

```python
data=data.reindex(columns=new_index)
```

```python
data
```

```python
stars=data['Star']
```

```python
stars.value_counts()
```

```python
for i in range(len(stars)):
    if int(stars[i])<=3:
        data.iloc[i,0]=0
    else:
        data.iloc[i,0]=1
```

```python
data
```

```python
data['label']=data['label'].astype('int64')
```

```python
data
```

```python
new_col=data.columns.drop(['Star'])
```

```python
data=data.reindex(columns=new_col)
```

```python
data['label'].value_counts()
```

```python
data
```

```python
data.to_csv("./new_big.csv",header=['label','Comment'],mode="a+",index=False,index_label=False)# 将处理好的数据重新保存至csv
```
