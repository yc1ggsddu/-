```python
import pandas as pd
```

```python
df1=pd.read_csv("./datasets/douban_Sniper.csv",encoding='utf-8',names=["comment_time","user_name","comment","rating","title"])
df2=pd.read_csv("./datasets/douban_Ganghaoxihuan.csv",encoding='utf-8',names=["comment_time","user_name","comment","rating","title"])
df3=pd.read_csv("./datasets/douban_niluohe.csv",encoding='utf-8',names=["comment_time","user_name","comment","rating","title"])
df4=pd.read_csv("./datasets/douban_shinian.csv",encoding='utf-8',names=["comment_time","user_name","comment","rating","title"])
df5=pd.read_csv("./datasets/douban_buyaowangji.csv",encoding='utf-8',names=["comment_time","user_name","comment","rating","title"])
df6=pd.read_csv("./datasets/douban_sihai.csv",encoding='utf-8',names=["comment_time","user_name","comment","rating","title"])
df7=pd.read_csv("./datasets/douban_shashou.csv",encoding='utf-8',names=["comment_time","user_name","comment","rating","title"])
```

```python
df_emp=pd.DataFrame()
```

```python
# 将上述得到的dataframe数据集进行合并
df_emp=pd.concat([df1,df2,df3,df4,df5,df6,df7],ignore_index=True)
```

```python
df_emp.loc[:,"rating"].value_counts()
```

```python
# 重新保存数据集
df_emp.to_csv("./datasets/douban_Combine.csv",mode='a+',header=["comment_time","user_names","comments","rating","title"],index=False,index_label=False)
```
